```python
#set threshold for missing data
list(newdf.dropna(thresh=int(newdf.shape[0] * .9), axis=1).columns)
```

```python
#missing info
missinginfo = list(newdf.columns[newdf.isnull().any()])
```

```python
missinginfo
```

```python
#columns for missing data
print("count of missing data :")
for col in missinginfo:
 num_missing= newdf[newdf[col].isnull() == True].shape[0]     #using for loop to iterate over the dataframe

 print('number missing for column {}: {}'.format(col, num_missing))
```

```python
#percentage of missing data
for col in missinginfo:                                              #using for loop to iterate over the dataframe
 num_missing = newdf[newdf[col].isnull() == True].shape[0] 
 print('number missing for column {}: {}'.format(col, num_missing)) #count of missing data
for col in missinginfo:
 percent_missing = newdf[newdf[col].isnull() == True].shape[0] /newdf.shape[0] 
 print('percent missing for column {}: {}'.format( col,percent_missing))         #percentage of missing data
```
